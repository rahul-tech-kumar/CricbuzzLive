```python
import requests

url = "https://cricbuzz-cricket.p.rapidapi.com/mcenter/v1/40381/hscard"

headers = {
	"x-rapidapi-key": "d86ff93035msh5d9cc5b9a4e3a07p1ea55ejsnfb047b8ba175",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

match_scard=response.json()
print(match_scard)
```

```python
import pandas as pd

score_cards = match_scard.get("scoreCard", [])

matches_batsman_list = []
matches_bowler_list = []

for card in score_cards:
    match_id = card.get("matchId")
    innings_id = card.get("inningsId")
    
    # ---------- Batting Team ----------
    bat_team = card.get("batTeamDetails", {})
    bat_team_id = bat_team.get("batTeamId")
    bat_team_name = bat_team.get("batTeamName")
    bat_team_short = bat_team.get("batTeamShortName")

    # ---------- Batsmen Data ----------
    batsmen_data = bat_team.get("batsmenData", {})
    for _, bat_info in batsmen_data.items():
        matches_batsman_list.append({
            "match_id": match_id,
            "innings_id": innings_id,
            "bat_team_id": bat_team_id,
            "bat_team_name": bat_team_name,
            "bat_team_short": bat_team_short,
            "bat_id": bat_info.get("batId"),
            "bat_name": bat_info.get("batName"),
            "is_captain": bat_info.get("isCaptain"),
            "is_keeper": bat_info.get("isKeeper"),
            "runs": bat_info.get("runs"),
            "balls": bat_info.get("balls"),
            "fours": bat_info.get("fours"),
            "sixes": bat_info.get("sixes"),
            "strike_rate": bat_info.get("strikeRate"),
            "out_desc": bat_info.get("outDesc"),
            "bowler_id": bat_info.get("bowlerId"),
            "fielder1_id": bat_info.get("fielderId1"),
            "fielder2_id": bat_info.get("fielderId2"),
            "fielder3_id": bat_info.get("fielderId3"),
            "wicket_code": bat_info.get("wicketCode")
        })

    # ---------- Bowling Team ----------
    bowl_team = card.get("bowlTeamDetails", {})
    bowl_team_id = bowl_team.get("bowlTeamId")
    bowl_team_name = bowl_team.get("bowlTeamName")
    bowl_team_short = bowl_team.get("bowlTeamShortName")

    # ---------- Bowlers Data ----------
    bowlers_data = bowl_team.get("bowlersData", {})
    for _, bowl_info in bowlers_data.items():
        matches_bowler_list.append({
            "match_id": match_id,
            "innings_id": innings_id,
            "bowl_team_id": bowl_team_id,
            "bowl_team_name": bowl_team_name,
            "bowl_team_short": bowl_team_short,
            "bowler_id": bowl_info.get("bowlerId"),
            "bowler_name": bowl_info.get("bowlerName"),
            "overs": bowl_info.get("overs"),
            "maidens": bowl_info.get("maidens"),
            "runs_conceded": bowl_info.get("runs"),
            "wickets": bowl_info.get("wickets"),
            "economy": bowl_info.get("economy"),
            "dots": bowl_info.get("dots"),
            "fours_conceded": bowl_info.get("fours"),
            "sixes_conceded": bowl_info.get("sixes")
        })

# Convert into DataFrames
df_batsmen = pd.DataFrame(matches_batsman_list)
df_bowlers = pd.DataFrame(matches_bowler_list)

# Save to CSV
df_batsmen.to_csv("matches_batsman_list.csv", index=False)
df_bowlers.to_csv("matches_bowler_list.csv", index=False)

print("Batsmen:\n", df_batsmen.head(), "\n")
print("Bowlers:\n", df_bowlers.head())
```
